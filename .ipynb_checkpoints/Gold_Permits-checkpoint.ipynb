```python
import sqlite3
import pandas as pd
from pathlib import Path

# Paths
PROJECT_ROOT = Path(r"C:\Users\sdnea\Dev\permit-data-dev")
DB_PATH = PROJECT_ROOT / "permit_data.db"

# Connect to SQLite
conn = sqlite3.connect(DB_PATH)
```

```python
# Drop empty columns & rename for presentation
query = """
DROP TABLE IF EXISTS permits_gold;

CREATE TABLE permits_gold AS
SELECT
    record_id AS "Record ID",
    record_type AS "Record Type",
    title AS "Title",
    permit_type AS "Permit Type",
    date_issued AS "Permit Date",
    type_of_property AS "Property Type",
    address AS "Address",
    valuation AS "Valuation ($)",
    contractor AS "Contractor",
    description AS "Description"
FROM permits_silver;
"""

conn.executescript(query)
conn.commit()
print("✅ permits_gold created")
```

```python
df_gold = pd.read_sql("SELECT * FROM permits_gold LIMIT 10;", conn)
df_gold
```

```python
OUTPUT_PATH = PROJECT_ROOT / "permits_gold.xlsx"

# Load the full gold table
df_gold_full = pd.read_sql("SELECT * FROM permits_gold;", conn)

# Export all rows
df_gold_full.to_excel(OUTPUT_PATH, index=False)
print(f"✅ Exported {len(df_gold_full)} rows to {OUTPUT_PATH}")
```
